```python
user_movies_data=pd.pivot_table(use,index='movieId',columns='userId',values='rating',fill_value=0)
user_movies_data
```

```python
movies_df['tags']=movies_df['tags'].fillna('None')
movies_df
```

```python
#Content Filtering
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
tfidf_matrix=tfidf.fit_transform(movies_df['tags'])
tfidf_df=pd.DataFrame(tfidf_matrix.toarray(),index=movies_df.index.tolist())
tfidf_df
```

```python
# Dimensionality Reduction using TruncatedSVD
from sklearn.decomposition import TruncatedSVD

# Define the number of components for TruncatedSVD
n_components = 19

# Initialize TruncatedSVD
svd = TruncatedSVD(n_components=n_components)

# Fit and transform the TF-IDF matrix to obtain the latent matrix
latent_matrix = svd.fit_transform(tfidf_df)


# Creating a DataFrame for Latent Matrix 1
n = 19  # Number of components to include
latent_matrix_1_df = pd.DataFrame(latent_matrix[:, :n], index=movies_df['title'].tolist())
```

```python
# Collaborative Filtering

svd = TruncatedSVD(n_components=20)
latent_matrix_2 = svd.fit_transform(user_movies_data)

latent_matrix_2_df = pd.DataFrame(latent_matrix_2, index=[movies_df.loc[movies_df.movieId == i, 'title'].values[0] for i in use['movieId'].unique()])

latent_matrix_2_df
```

```python
#Finding issues with different sized matrixes
latent_matrix_1_df=latent_matrix_1_df.drop_duplicates()
latent_matrix_2_df=latent_matrix_2_df.drop_duplicates()
```

```python
a=latent_matrix_2_df.copy()
for i in latent_matrix_1_df.index:
    if i not in latent_matrix_2_df.index:
        a=a.append(pd.Series(np.zeros(20), index=np.arange(0,20),name=i))
```

```python
b=latent_matrix_1_df.copy()
for i in a.index:
    if i not in b.index:
        b=b.append(pd.Series(np.zeros(19), index=np.arange(0,19),name=i))
```

```python
#Same index was repeated multiple times, decided to remove duplicates
a = a[~a.index.duplicated(keep='first')]
```

```python
b = b[~b.index.duplicated(keep='first')]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar_movies(title):
    if title in b.index:
        a_1=np.array(b.loc[title]).reshape(1,-1)
        score_content=cosine_similarity(b,a_1).reshape(-1)
    else:
        score_content=0
    if title in a.index:
        a_2=np.array(a.loc[title]).reshape(1,-1)
        score_collab=cosine_similarity(a,a_2).reshape(-1)
    else:
        score_collab=0

    hybrid_score=(score_content+score_collab)/2

    dictDF={'content':score_content,'collab':score_collab,'hybrid':hybrid_score}
    dictDF
    similar_movies=pd.DataFrame(dictDF,index=a.index)


    similar_movies.sort_values('hybrid',ascending=False,inplace=True)
    return similar_movies
```

```python
recommend_similar_movies('Shine a Light (2008)')
```

```python
recommend_similar_movies('Interview with the Vampire: The Vampire Chronicles (1994)')
```

```python
from surprise import Dataset, Reader, SVD, accuracy
from surprise.model_selection import train_test_split
Mapping_file = dict(zip(movies_df['title'].tolist(), movies_df['movieId'].tolist()))

reader=Reader(rating_scale=(1,5))
data=Dataset.load_from_df(use[['userId','movieId','rating']],reader)

#using split data 'use' from earlier
trainset,testset=train_test_split(data,test_size=0.25)

svd=SVD()
svd.fit(trainset)
preds=svd.test(testset)

accuracy.rmse(preds)
```

```python
def predict_user_ratings(user_id):
    if user_id in ratings_df.userId.unique():
        user_movie_ids = ratings_df[ratings_df.userId == user_id].movieId.tolist()
        movie_id_map = {k: v for k, v in Mapping_file.items() if v not in user_movie_ids}
        predicted_list = []
        for movie_id, mapped_id in movie_id_map.items():
            predicted_rating = svd.predict(user_id, mapped_id)
            predicted_list.append((movie_id, predicted_rating[3]))
        predicted_df = pd.DataFrame(predicted_list, columns=['movies', 'ratings'])
        predicted_df.sort_values('ratings', ascending=False, inplace=True)
        predicted_df.set_index('movies', inplace=True)
        return predicted_df.head(10)
    else:
        print("User ID does not exist in the list!")
        return None
```
